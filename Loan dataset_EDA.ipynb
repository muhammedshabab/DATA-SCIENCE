# Data Pre-Processing

### a) Missing values handling

```python
loan_df.isnull().sum()
```

```python
loan_df.dtypes
```

```python
loan_df["past_due_days"]=loan_df["past_due_days"].fillna(loan_df['past_due_days'].median())
```

```python
loan_df.isnull().sum()
```

```python
loan_df["paid_off_time"]=loan_df["paid_off_time"].fillna(loan_df['paid_off_time'].mode()[0])
```

```python
loan_df.isnull().sum()
```

```python
loan_df.loc[(loan_df['loan_status']=='Paidoff')&(loan_df['past_due_days'].isnull()),'past_due_days']=0
```

```python
loan_df.groupby('education')['age'].mean()
```

###### The average past due days, loan terms, and borrower age vary by education level. Generally, borrowers with higher education tend to have fewer overdue days and shorter loan terms, indicating better repayment behavior and possibly different borrowing needs. Age differences reflect typical demographic patterns within each education group.

```python
loan_df.groupby('education')[['past_due_days','terms','past_due_days','age']].mean()
```

### b) Outlier detection and handling 

### histplot

```python
sns.histplot(loan_df['past_due_days'], kde=True)
```

##### The histogram shows most borrowers have low past due days, with a right-skewed distributionfew borrowers have very high overdue days. The KDE line highlights this concentration around lower values.

```python
num_cols = ['Principal','terms','past_due_days','age']
for col in num_cols:
    plt.figure()
    loan_df.boxplot(column=[col])
    plt.title(f"Before Capping: {col}")
    plt.show()
```

```python
for col in num_cols:
    Q1=loan_df[col].quantile(0.25)
    Q3=loan_df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_cap=Q1-1.5*IQR
    upper_cap=Q3+1.5*IQR
    loan_df[col]=loan_df[col].clip(lower_cap,upper_cap)
```

```python
for col in num_cols:
    plt.figure()
    loan_df.boxplot(column=[col])
    plt.title(f"after capping:{col}")
    plt.show()
```
